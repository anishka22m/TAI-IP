### Visualizations and Insights

```python
#How Many employees have left the Company 
#Encoded - 0:No, 1:Yes
sns.countplot(data=df, x='Attrition', palette='Set1')
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
```

## Observation 
Not a lot Attrition of has been observed

```python
#department genderwise attrition
#Encoded :  'Sales':1,'Research & Development':2,'Human Resources':3
##Encoded : 'Female':0,'Male':1
sns.countplot(data=df, x='Department', hue='Gender')
plt.title('Attrition by Department and Gender')
plt.xlabel('Department')
plt.ylabel('Count')
```

## Observation 
1. Most Employee Attrition is from Research and Development Department
2. More males have left than females in R&D
3. Least Employee Attrition is from Human Resources Department
4. More males have left than females in Hr

```python
#Dailyrate vs Attrition
#Encoded - 0:No, 1:Yes
sns.barplot(x=df['Attrition'],y=df['DailyRate'])
plt.title('Attrition Vs DailyRate')
plt.xlabel('Attrition')
plt.ylabel('DailyRate')
```

## Observation
Higher Daily rate has lower Attrition as employees with Higher dailyrate have left the company less

```python
df.head()
```

```python
#Attrition vs Years Since Last Promotion and Years in Current Role
#Encoded - Attrition: - 0:No, 1:Yes
sns.scatterplot(data=df, x='YearsSinceLastPromotion', y='YearsInCurrentRole', hue='Attrition', palette='Set1')
plt.title('Attrition vs Years Since Last Promotion and Years in Current Role')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Years in Current Role')
plt.legend(title='Attrition')
plt.show()
```

## Observation 
Employees who do no have recent promotions and have been in the same role for longer time tend to leave more

```python
sns.barplot(data=df, x='Attrition', y='EnvironmentSatisfaction', palette='Set1')
plt.title('Environment Satisfaction by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Average Environment Satisfaction')
plt.show()
```

```python
#percentSalaryHike genderwise attrition
##Encoded : 'Female':0,'Male':1
sns.countplot(data=df, x='PercentSalaryHike', hue='Gender')
plt.title('Attrition by Department and Gender')
plt.xlabel('Percent Salary Hike')
plt.ylabel('Count')
```

## Observation
Males have had relatively higher number of PercentSalary Hike

```python
##Encoded : 'Female':0,'Male':1
sns.barplot(data=df, x='Gender', y='Attrition', palette='Set1')
plt.title('Gender vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Gender')
plt.show()
```

## Observation
Even though women have not received percent salary hike as high in number as men yet attrition is higher for men

```python
atcomp_proportion = df.groupby('YearsAtCompany')['Attrition'].sum() / df['YearsAtCompany'].value_counts()
atrole_proportion = df.groupby('YearsInCurrentRole')['Attrition'].sum() / df['YearsInCurrentRole'].value_counts()
plt.figure(figsize=(10, 6))
plt.plot(atcomp_proportion.index, atcomp_proportion.values, marker='o', linestyle='-',color='red',label='YearsAtCompany')
plt.plot(atrole_proportion.index, atrole_proportion.values, marker='o', linestyle='-',color='blue',label='YearsInCurrentRole')
plt.title('Attrition Proportion by Years at Company')
plt.xlabel('Years at Company/Years in Current Role')
plt.ylabel('Attrition Proportion')
plt.legend()
plt.grid(True)
plt.show()
```

## Observation
1. Maximum attrition is after 35 years in in the company 
2. Attrition Hike is there after approximately 10-12 years in the current Role

```python
df['WorkLifeBalance'].unique()
```

```python
worklife_ratings = df.groupby('WorkLifeBalance')['Attrition'].mean().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(worklife_ratings['WorkLifeBalance'], worklife_ratings['Attrition'], color='skyblue')
plt.title('Attrition Rate by Work-Life Balance Rating')
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Average Attrition Rate')
plt.xticks(range(1, 5), ['Bad', 'Good', 'Better', 'Best'])
plt.show()
```

## Observation
We Determine an immportance of Work-Life Balance

```python
###ncoded - Female:0,Male:1
##Encoded - 'Travel_Rarely':1,'Travel_Frequently':2,'Non-Travel':3
attrition_rate_by_travel_gender = df.groupby(['BusinessTravel', 'Gender'])['Attrition'].mean().reset_index()
plt.figure(figsize=(10, 6))
# Create a list of x-axis labels for each combination
labels = [f'{travel} - {gender}' for travel, gender in zip(attrition_rate_by_travel_gender['BusinessTravel'], attrition_rate_by_travel_gender['Gender'])]
plt.bar(labels, attrition_rate_by_travel_gender['Attrition'], color=['skyblue', 'orange', 'green'])
plt.title('Attrition Rate by Business Travel and Gender')
plt.xlabel('Business Travel - Gender')
plt.ylabel('Average Attrition Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Observation
We Determine that Women who travel more frequently have the maximum number of Avg Attrition rate followed by Men who travel Frequently

```python

attrition_rate = df.groupby('PerformanceRating')['Attrition'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(attrition_rate['PerformanceRating'], attrition_rate['Attrition'], color='pink',width=0.3)
plt.title('Average Attrition Rate by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Average Attrition Rate')
plt.xticks(attrition_rate['PerformanceRating'])  # Set x-axis labels
# plt.tight_layout()
plt.ylim(0, 0.2)
plt.show()
```

## Observation 
The attrition rate is almost same for the performance rating
